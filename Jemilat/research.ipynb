```python
#- Is there a correlation between the type of incident and the number of deaths?

st.subheader("Proportion of Incidents per State")
state_counts = df["State"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%')
st.pyplot()

 #Which states experience the widest range of incident types?

st.subheader("Distribution of Deaths")
plt.figure(figsize=(8,6))
sns.histplot(df["Number of deaths"], bins=20, kde=True)
st.pyplot()


#  - Which states report the highest number of deaths across incidents?

st.subheader("Incidents by State")
fig1 = sns.catplot(x="State", kind="count", data=df, height=8, aspect=2, palette="Set2")
fig1.set_axis_labels("State", "Number of Incidents", fontsize=14)
fig1.set_xticklabels(rotation=45, fontsize=12)
fig1.set_yticklabels(fontsize=12)
fig1.fig.suptitle("Incidents by State", fontsize=18)
st.pyplot(fig1.fig)

# - How do incidents vary over time — are there seasonal or yearly trends?

st.subheader("Deaths Distribution by Incident Type")
fig2, ax2 = plt.subplots(figsize=(10,6))
sns.violinplot(x="Incident", y="Number of deaths", data=df, ax=ax2, palette="muted")
ax2.set_title("Deaths Distribution by Incident Type", fontsize=16)
ax2.set_xlabel("Incident", fontsize=14)
ax2.set_ylabel("Number of Deaths", fontsize=14)
st.pyplot(fig2)

# - Is there a relationship between incident type and number of deaths?

st.subheader("Proportion of Incidents per State")
state_counts = df["State"].value_counts()
fig3, ax3 = plt.subplots(figsize=(6,6))
ax3.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%', startangle=90)
ax3.set_title("Proportion of Incidents per State", fontsize=16)
st.pyplot(fig3)

#- Which states experience the widest variety of incident types?

st.subheader("Distribution of Deaths")
fig4, ax4 = plt.subplots(figsize=(8,6))
sns.histplot(df["Number of deaths"], bins=20, kde=True, ax=ax4, color="skyblue")
ax4.set_title("Distribution of Deaths", fontsize=16)
ax4.set_xlabel("Number of Deaths", fontsize=14)
ax4.set_ylabel("Frequency", fontsize=14)
st.pyplot(fig4)

#  What is the distribution of deaths per incident (low vs. high fatality)?

st.subheader("Deaths Over Time")
df["Start date"] = pd.to_datetime(df["Start date"])
fig5, ax5 = plt.subplots(figsize=(10,6))
sns.lineplot(x="Start date", y="Number of deaths", data=df, ax=ax5, marker="o")
ax5.set_title("Deaths Over Time", fontsize=16)
ax5.set_xlabel("Date", fontsize=14)
ax5.set_ylabel("Number of Deaths", fontsize=14)
st.pyplot(fig5)


 
```
